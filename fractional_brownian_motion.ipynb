Simulate Levy's fractional Brownian motion (fBm) paths $W^H$

By definition, we have

$$
W_t^H = \sqrt{2H} \int_0^t (t-s)^{H-1/2} dB_s,
$$

where $H \in (0,1)$ is the Hurst exponent.

```python
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from stochastic.fbm import simulate_fbm

sns.set_theme("talk")
mpl.rcParams["figure.figsize"] = (8, 6)
```

### Compare fBm paths for different values of the Hurst exponent

```python
SEED = 1234  # set seed for reproducibility
t = 0.5
Hs = [0.1, 0.3, 0.5, 0.7, 0.9]
n_steps = 300
n_paths = 1
```

```python
fig, ax = plt.subplots()
for H in Hs:
    tab_t, fbm_path = simulate_fbm(t, H, n_steps, n_paths, seed=SEED)
    ax.plot(tab_t, fbm_path, "-", label=f"$H={H}$")
ax.legend(ncol=3)
ax.set_xlabel("$t$")
ax.set_ylabel("$W_t^H$")
plt.show()
```
